```python
# Adama Ndiaye


class Livre:
    def __init__(self, id, titre, auteur, disponible=True):
        self.id = id
        self.titre = titre
        self.auteur = auteur
        self.disponible = disponible

class Auteur:
    def __init__(self, nom, nationalite):
        self.nom = nom
        self.nationalite = nationalite
        self.oeuvres = []  # Liste des oeuvres de l'auteur

    def ajouter_oeuvre(self, oeuvre):
        self.oeuvres.append(oeuvre)

class Emprunteur:
    def __init__(self, id, nom):
        self.id = id
        self.nom = nom
        self.livres_empruntes = []  # Liste des livres empruntés

class Bibliotheque:
    def __init__(self):
        self.livres = []
        self.auteurs = []
        self.emprunteurs = []

    def ajouter_livre(self, id, titre, auteur, disponible=True):
        # Vérifie si un livre avec le même ID existe déjà
        for livre in self.livres:
            if livre.id == id:
                print(f"Erreur : Un livre avec l'ID {id} existe déjà.")
                return False
        livre = Livre(id, titre, auteur, disponible)
        self.livres.append(livre)
        print(f"Livre '{titre}' de {auteur} ajouté à la bibliothèque.")
        return True

    def rechercher_livre(self, recherche):
        livres_trouves = []
        for livre in self.livres:
            if recherche.lower() in livre.titre.lower() or recherche.lower() in livre.auteur.lower():
                livres_trouves.append(livre)

        # Affiche les résultats de la recherche
        if livres_trouves:
            for livre in livres_trouves:
                status = "disponible" if livre.disponible else "emprunté"
                print(f"ID: {livre.id}, Titre: {livre.titre}, Auteur: {livre.auteur}, Status: {status}")
        else:
            print("Aucun livre trouvé pour cette recherche.")
        return livres_trouves

    def emprunter_livre(self, id_livre, id_emprunteur):
        livre = next((livre for livre in self.livres if livre.id == id_livre), None)
        emprunteur = next((e for e in self.emprunteurs if e.id == id_emprunteur), None)

        if not livre:
            print(f"Erreur : Le livre avec l'ID {id_livre} n'existe pas.")
            return False

        if not emprunteur:
            print(f"Erreur : L'emprunteur avec l'ID {id_emprunteur} n'existe pas.")
            return False

        if not livre.disponible:
            print(f"Erreur : Le livre '{livre.titre}' est déjà emprunté.")
            return False

        livre.disponible = False
        emprunteur.livres_empruntes.append(livre)
        print(f"Livre '{livre.titre}' emprunté à {emprunteur.nom}.")
        return True

    def retourner_livre(self, id_livre, id_emprunteur):
        # Recherche du livre
         livre = None
         for l in self.livres:
               if l.id == id_livre:
                livre = l
                break

        # Recherche de l'emprunteur
         emprunteur = None
         for e in self.emprunteurs:
          if e.id == id_emprunteur:
           emprunteur = e
           break

         if not livre:
            print(f"Erreur : Le livre avec l'ID {id_livre} n'existe pas.")
            return False

         if not emprunteur:
            print(f"Erreur : L'emprunteur avec l'ID {id_emprunteur} n'existe pas.")
            return False

         if livre not in emprunteur.livres_empruntes:
            print(f"Erreur : Le livre '{livre.titre}' n'a pas été emprunté par {emprunteur.nom}.")
            return False

         livre.disponible = True
         emprunteur.livres_empruntes.remove(livre)
         print(f"Livre '{livre.titre}' retourné par {emprunteur.nom}.")
         return True

    def ajouter_emprunteur(self, id, nom):
        emprunteur = Emprunteur(id, nom)
        self.emprunteurs.append(emprunteur)
        print(f"Emprunteur '{nom}' ajouté à la bibliothèque.")

# Exemple d'utilisation
bibliotheque = Bibliotheque()

# Ajout d'un auteur
auteur1 = Auteur("Birago Diop", "Sénégalais")
auteur1.ajouter_oeuvre("Cahier d'un retour au pays natal")
auteur1.ajouter_oeuvre("Les contes d'Amadou Koumba")
bibliotheque.auteurs.append(auteur1)

# Ajout de livres
bibliotheque.ajouter_livre(1, "Les contes d'Amadou Koumba", "Birago Diop")
bibliotheque.ajouter_livre(2, "Cahier d'un retour au pays natal", "Birago Diop")

# Ajout d'un emprunteur
emprunteur1 = Emprunteur(44, "Adama")
bibliotheque.emprunteurs.append(emprunteur1)

# Rechercher un livre
bibliotheque.rechercher_livre("Les contes d'Amadou Koumba")

# Emprunt et retour de livre
bibliotheque.emprunter_livre(1, 44)
bibliotheque.rechercher_livre("Cahier d'un retour au pays natal")  # Vérifier l'état après emprunt
bibliotheque.retourner_livre(1, 44)
bibliotheque.rechercher_livre("Cahier d'un retour au pays natal")  # Vérifier l'état après retour
```

```python
import unittest

# Définition de la bibliothèque et des classes Livre, Emprunteur, etc.
# Assurez-vous que les classes Livre, Emprunteur, Bibliotheque sont bien définies avant d'exécuter ce test.

class TestBibliotheque(unittest.TestCase):
    def setUp(self):
        self.bibliotheque = Bibliotheque()

        # Creer Livre
        self.livre1 = Livre(1, "L'enfant Noir", "Camara Laye")
        self.livre2 = Livre(2, "L'aventure ambigue", "Cheikh Hamidou Kane")
        self.livre3 = Livre(3, "Les contes d'Amadou Koumba", "Birago Diop")
        # Creer emprunteur
        self.emprunteur = Emprunteur(44, "Adama Ndiaye")
        self.emprunteur2 = Emprunteur(45, "Cheikh Diop")

        self.bibliotheque.livres = [self.livre1, self.livre2, self.livre3]
        self.bibliotheque.emprunteurs = [self.emprunteur, self.emprunteur2]  # Ajout des emprunteurs à la bibliothèque

    def test_ajouter_livre(self):

        # Cas positif
        ajout = self.bibliotheque.ajouter_livre(4, "Cahier d'un retour au pays natal", "Birago diop")
        self.assertTrue(ajout)
        self.assertEqual(len(self.bibliotheque.livres), 4)

        # Cas négatif (livre déjà existant)
        ajout = self.bibliotheque.ajouter_livre(4, "Cahier d'un retour au pays natal", "Birago diop")
        self.assertFalse(ajout)
        self.assertEqual(len(self.bibliotheque.livres), 4)

    def test_rechercher_livre(self):

        # Test cas positif : recherche par titre
        result = self.bibliotheque.rechercher_livre("Les contes d'Amadou Koumba")
        print(f"Recherche par titre: {result}")
        self.assertEqual(len(result), 1)
        self.assertEqual(result[0].titre, "Les contes d'Amadou Koumba")

        # Test cas positif : recherche par auteur
        result = self.bibliotheque.rechercher_livre("Birago diop")
        print(f"Recherche par auteur: {result}")
        self.assertEqual(len(result), 1)

        # Test cas négatif : recherche sans résultat
        result = self.bibliotheque.rechercher_livre("Inexistant")
        print(f"Recherche sans résultat: {result}")
        self.assertEqual(len(result), 0)


    def test_emprunter_livre(self):

        # Test cas positif : emprunt réussi
        result = self.bibliotheque.emprunter_livre(1, 44)
        self.assertTrue(result)
        self.assertFalse(self.bibliotheque.livres[0].disponible)
        self.assertIn(self.livre1, self.emprunteur.livres_empruntes)

        # Test cas négatif : emprunt d'un livre déjà emprunté
        result = self.bibliotheque.emprunter_livre(1, 44)
        self.assertFalse(result)

        # Test cas négatif : emprunt d'un livre inexistant
        result = self.bibliotheque.emprunter_livre(99, 44)
        self.assertFalse(result)

        # Test cas négatif : emprunt par un emprunteur inexistant
        result = self.bibliotheque.emprunter_livre(1, 99)
        self.assertFalse(result)

    def test_retourner_livre(self):

        # Emprunter le livre d'abord pour tester le retour
        self.bibliotheque.emprunter_livre(1, 44)

        # Test cas positif : retour réussi
        result = self.bibliotheque.retourner_livre(1, 44)
        self.assertTrue(result)
        self.assertTrue(self.bibliotheque.livres[0].disponible)
        self.assertNotIn(self.livre1, self.emprunteur.livres_empruntes)

        # Test cas négatif : retour d'un livre non emprunté
        result = self.bibliotheque.retourner_livre(2, 44)
        self.assertFalse(result)

        # Test cas négatif : retour par un emprunteur inexistant
        result = self.bibliotheque.retourner_livre(1, 99)
        self.assertFalse(result)

        # Test cas négatif : livre inexistant
        result = self.bibliotheque.retourner_livre(99, 44)
        self.assertFalse(result)

if __name__ == "__main__":

    # Chargement des tests
    test_loader = unittest.TestLoader()  # Déclare une variable "Loader"
    test_suite = test_loader.loadTestsFromTestCase(TestBibliotheque)  # Charge les tests
    test_runner = unittest.TextTestRunner(verbosity=2)
    test_runner.run(test_suite)  # Exécute les tests chargés
```
